*RGBA ---> o RGB repesentam as cores Red, Green e Blue, todas varu
alpha representa a transparencia dessas cores, sendo 0 transparente e 1 a coloração máxima

### Grafico de Barras

```python
import numpy as np
import matplotlib.pyplot as plt

categorias = ["A", "B", "C", "D"]
valores = [10, 15, 7, 12]

plt.figure()
plt.barh(categorias, valores, color = "blue", alpha = 0.75)
plt.xlabel = ("categorias")
plt.ylabel = ("valores")
plt.title("* * * GRAFICO DE BARRAS * * *")
plt.show()
```

###Histograma

```python
import numpy as np
import matplotlib.pyplot as plt

dados = np.random.randn(1000) #teste
plt.figure()
plt.hist(dados, bins = 30, color = "purple", alpha = 0.75)
plt.xlabel = ("categorias")
plt.ylabel = ("valores")
plt.title("* * * HISTOGRAMA * * *")
plt.show()
```
